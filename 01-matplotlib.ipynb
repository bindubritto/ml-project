```python
from matplotlib import pyplot as plt
# Note: The pyplot module is a simple interface for matplotlib
```

```python
plt.plot([1, 2, 3], [1, 4, 9])
plt.plot([1, 2, 3], [10, 20, 30])
plt.legend(['Data Set 01', 'Data Set 02'])
# plt.legend() takes a list as an argument

plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.title('This is a line chart')

#plt.show()

# To save image using plt.savefig(), we need to comment plt.show()
plt.savefig('line_chart')
```
